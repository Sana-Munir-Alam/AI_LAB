```python
# Task 1:
import random

class Environment:
    def __init__(self):
        self.components = {}
        for i in range(9):
            self.components[chr(ord('A') + i)] = random.choice(['Safe', 'Vulnerable'])
    
    def displaySystem(self, title):
        print(f"\n{title}")
        for comp, status in self.components.items():
            print(f"Component {comp}: {status}")
    
    def get_percept(self, component):
        return self.components[component]
    
    def patch(self, component):
        self.components[component] = 'Safe'


class SecurityAgent:
    def __init__(self):
        self.to_patch = []
    
    def scan(self, component, percept):         # Agent Acting
        if percept == 'Vulnerable':
            print(f"Warning: Component {component} is VULNERABLE. Adding to list for PATCHING")
            self.to_patch.append(component)
        else:
            print(f"Component {component} is secure")
    
    def patch_vulnerabilities(self, environment):
        print("\nPatching Vulnerabilities...")
        for comp in self.to_patch:
            environment.patch(comp)
            print(f"Component {comp} patched successfully")
    
    def run_agent(self, environment):
        environment.displaySystem("Initial System State")
        
        # System Scan
        print("\nScanning System...")
        for component in environment.components:
            percept = environment.get_percept(component)
            self.scan(component, percept)
        
        # Patching
        self.patch_vulnerabilities(environment)
        
        # Final System Check
        environment.displaySystem("Final System State")


# MAIN SECTION
agent = SecurityAgent()
environment = Environment()
agent.run_agent(environment)
```

```python
# Task 2
import random

class Environment:
    def __init__(self):
        self.servers = {}
        for i in range(5):  # for 5 server state their service
            self.servers[i+1] = random.choice(['Underloaded', 'Balanced', 'Overloaded'])

    def display(self, title):
        print(f"\n{title}")
        for server, load in self.servers.items():
            print(f"{server}: {load}")

    def get_percept(self):
        return self.servers             # returns everything for the the whole system is being scanned so Agent will be aware of everything

    def update_server(self, server, new_state):
        self.servers[server] = new_state


class LoadBalancerAgent:
    def __init__(self):
        self.model = {}                 # Model-Based (store internal memory on where task is overloaded, and underlaoded so that it can later maintain balance)

    def update_model(self, percept):    # Update model memory with what it is percieving
        self.model = percept.copy()     # Creates a new dictionary to Prevent accidental modification of environment

    def balance_load(self, environment):
        overloaded = []     # List to maintain which server are over laoded so that we can take task OUT from them to balance them
        underloaded = []    # List to maintain which server are under loaded so that we can put task IN them to balance them

        for server, state in self.model.items():        # Looping through agent internal memory (tuple format where server is (1-5) state is Balanced/Underloaded/Overloaded)
            if state == "Overloaded":                   # if server state is Over loaded add to list
                overloaded.append(server)               # store server index
            elif state == "Underloaded":
                underloaded.append(server)
            # else server is balanced do nothing

        print("\n Balancing Load...")
        
        while overloaded and underloaded:               # loop if we have both equal amount of loaded and underlaoded server
            o = overloaded.pop()                        # Get overloaded server index
            u = underloaded.pop()                       # Get underloaded server index
            if o == u:                                  # Same server edge case (though wont happen with 5 server but still doing it as a safety check3)
                print(f"Server {o} can't balance with itself")
                continue                                # skip this iteration

            print(f"Moving tasks from {o} â†’ {u}")
            environment.update_server(o, "Balanced")    # Now the overloaded server has it's task moved so it's balanced
            environment.update_server(u, "Balanced")    # Now the underloaded server has task added so it's balanced

        # After while loop, handle leftovers if there is
        for server in overloaded:
            print(f"Server {server} still overloaded (no underloaded server to help)")
        for server in underloaded:
            print(f"Server {server} still underloaded (no overloaded server to take from)")

    def act(self, percept, environment):
        self.update_model(percept)
        self.balance_load(environment)

    def run_agent(self, environment):
        environment.display("Initial Server Load Status")       # Initial system state
        percept = environment.get_percept()                     # Agent perception and action
        self.act(percept, environment)
        environment.display("Final Server Load Status")         # Final system state


# Main Section
env = Environment()
agent = LoadBalancerAgent()
agent.run_agent(env)

# one problem is that we have more overlaoded server then underloaded one. in this case we will still have server
# with overloaded task...as system doesnt have any underloaded servers to transfer the task to.
# so i am printing this via for loop that we have left over eith Overlaoded server or under loaded server.
```

```python
# Task 3
import random

class Environment:
    def __init__(self):
        self.Task = {}      # Tasks that are either 'Failed' or 'Completed'
        for i in range(5):  # Assuming 5 tasks
            self.Task[i + 1] = random.choice(['Failed', 'Completed'])

    def get_percept(self, index):
        return self.Task[index]

    def updateTask(self, index, new_state):  # Update the task with new status
        self.Task[index] = new_state

    def display(self, title):
        print(f"\n{title}")
        for task, status in self.Task.items():
            print(f"Task {task}: {status}")
        print("\n")

class BackupManagementAgent:
    def __init__(self):
        pass  # No internal state needed for simple reflex

    def act(self, percept):
        if percept == 'Failed':
            return 'Completed'      # Assuming that on teh first retry the backup is successful
        else:
            return 'Task Already Successfully Completed'

    # run_agent is now a method of the Agent
    def run_agent(self, environment):
        environment.display('Initial Task Status')

        for i in range(1, 6):  # Task indices from 1 to 5
            percept = environment.get_percept(i)
            action = self.act(percept)
            print(f"Step {i}: Percept: {percept}, Action: {action}")

            if action == 'Completed':
                environment.updateTask(i, action)

        environment.display('Final Task Status')


# -------- MAIN --------
environment = Environment()
agent = BackupManagementAgent()
agent.run_agent(environment)
```

```python
# Task 4

import random

class Environment:
    def __init__(self):
        self.components = {}
        for i in range(9):
            self.components[chr(ord('A') + i)] = random.choice(['Safe', 'Low Risk Vulnerable', 'High Risk Vulnerable'])

    def get_percept(self):
        return self.components.copy()

    def display(self, title):
        print(f"\n{title}")
        for comp, status in self.components.items():
            print(f"Component {comp}: {status}")

    def update_component(self, component, new_state):
        self.components[component] = new_state


class CyberSecurityAgent:
    def __init__(self):
        self.utility_scores = {'Safe': 0, 'Low Risk Vulnerable': 10, 'High Risk Vulnerable': -5}

    def utility(self, state):
        return self.utility_scores[state]

    def act(self, percept):
        best_component = None
        best_utility = -float('inf')

        for comp, state in percept.items():
            score = self.utility(state)
            if score > best_utility:
                best_utility = score
                best_component = comp

        state = percept[best_component]

        if state == 'Low Risk Vulnerable':
            action = 'Patch'
        elif state == 'High Risk Vulnerable':
            action = 'Requires Premium Service'
        else:
            action = 'No Action Needed'

        return best_component, action

    def run_agent(self, environment):
        environment.display("Initial System Status")
        print("")

        # First pass: Patch all Low Risk Vulnerable components
        while True:
            percept = environment.get_percept()
            if all(state != 'Low Risk Vulnerable' for state in percept.values()):
                break

            component, action = self.act(percept)
            state = percept[component]

            if state == 'Low Risk Vulnerable':
                print(f"Warning: Component {component} is LOW RISK VULNERABLE. Action: {action}.")
                environment.update_component(component, 'Safe')
                print(f"Component {component} patched successfully.")
            elif state == 'Safe':
                print(f"Component {component} is SAFE. No action needed.")
            elif state == 'High Risk Vulnerable':
                print(f"Warning: Component {component} is HIGH RISK VULNERABLE. Action: {action}")

        print("\nLow Risk Vulnerability HANDLED...now High Risk Vulnerability being checked\n")
        # Second pass: Log all High Risk components without patching
        percept = environment.get_percept()
        for component, state in percept.items():
            if state == 'High Risk Vulnerable':
                print(f"Warning: Component {component} is HIGH RISK VULNERABLE. Action: {action}")
            elif state == 'Safe':
                print(f"Component {component} is SAFE. No action needed.")

        environment.display("Final System Status")

# MAIN SECTION
env = Environment()
agent = CyberSecurityAgent()
agent.run_agent(env)
```

```python
# Task 5:

class Environment:
    def __init__(self):
        # Format: room_number: (medicine_type, patient_id, staff_available, delivered, patient_verified)
        self.rooms = {
            '101': ('Painkiller', 'P001', True, False, False),
            '102': ('Antibiotic', 'P002', False, False, False),
            '103': ('Vitamin', 'P003', True, False, False),
            '104': ('Insulin', 'P004', True, False, False),
            '105': ('Cough Syrup', 'P005', False, False, False)
        }
        self.storage = {'Painkiller': 5, 'Antibiotic': 5, 'Vitamin': 5, 'Insulin': 5, 'Cough Syrup': 5}
        self.robot_inventory = []
        self.delivery_log = []
    
    def get_percept(self, room):
        return self.rooms.get(room, None)
    
    def pick_medicine(self, medicine):
        if self.storage.get(medicine, 0) > 0:
            self.storage[medicine] -= 1
            self.robot_inventory.append(medicine)
            return True
        return False
    
    def verify_patient(self, room):
        medicine, patient_id, staff_available, delivered, already_verified = self.rooms[room]   # Returns True if patient ID is verified, False if needs staff alert
        if already_verified:
            return True, f"Patient {patient_id} already verified"
        elif staff_available:                                                                   # Mark as verified if Staff Exist
            self.rooms[room] = (medicine, patient_id, staff_available, delivered, True)
            return True, f"Patient {patient_id} verified successfully"
        else:
            return False, f"Staff not available to verify patient {patient_id}"
    
    def deliver_medicine(self, room):
        medicine, patient_id, staff_available, already_delivered, patient_verified = self.rooms[room]
        if already_delivered:
            return False, f"Medicine already delivered to room {room}"
        
        if medicine in self.robot_inventory and patient_verified:
            self.robot_inventory.remove(medicine)                                               # Remove from inventory
            self.rooms[room] = (medicine, patient_id, staff_available, True, patient_verified)  # Mark as delivered
            self.delivery_log.append(f"Room {room}: {medicine} delivered to {patient_id}")      # Add in Deliverd Log
            return True, f"Medicine {medicine} delivered to patient {patient_id} in room {room}"
        elif medicine not in self.robot_inventory:
            return False, f"Medicine {medicine} not in robot inventory"
        else:                                                                                   # Patient not verified
            return False, f"Cannot deliver - patient {patient_id} not verified"
    
    def alert_staff(self, room):
        _, _, staff_available, _, _ = self.rooms[room]
        if not staff_available:
            return f"Alert: Staff required at room {room} for patient verification"
        return f"Staff available at room {room}"
    
    def display_state(self, title):
        print(f"\n{title}")
        print("-" * 70)
        for room, (medicine, patient_id, staff_available, delivered, verified) in self.rooms.items():
            delivered_status = "Delivered" if delivered else "Pending"
            staff_status = "Available" if staff_available else "Not Available"
            verified_status = "Verified" if verified else "Not Verified"
            print(f"Room {room}: {medicine} for {patient_id} | Staff: {staff_status} | Patient: {verified_status} | Delivery: {delivered_status}")


class GoalBasedAgent:
    def __init__(self, environment):
        self.env = environment
        self.goal = "deliver_all_medicines"
    
    def run_agent(self):
        self.env.display_state("INITIAL HOSPITAL STATE")
        
        print("\n" + "=" * 70)
        print("ROBOT AGENT EXECUTION - DELIVERING MEDICINES")
        print("=" * 70)
        
        # Process each room
        for step, room in enumerate(self.env.rooms.keys(), 1):
            print(f"\n{'='*40}")
            print(f"STEP {step}: PROCESSING ROOM {room}")
            print(f"{'='*40}")
            
            room_data = self.env.get_percept(room)      # Get room information
            if not room_data:
                print(f"Room {room} not found!")
                continue
            
            medicine, patient_id, staff_available, delivered, verified = room_data
            
            # Skip if already delivered
            if delivered:
                print(f"Room {room}: Medicine already delivered")
                continue
            
            print(f"Required: {medicine} for patient {patient_id}")
            
            # 1. Go to storage and pick medicine
            print(f"1. Going to storage...")
            if self.env.pick_medicine(medicine):
                print(f"   Picked {medicine} from storage")
                print(f"   Robot inventory: {self.env.robot_inventory}")    # Print whole robot inventory
            else:
                print(f"   Failed to pick {medicine} - out of stock")
                continue
            
            # 2. Go to the room
            print(f"2. Moving to room {room}...")
            print(f"   Arrived at room {room}")
            
            # 3. Verify patient ID
            print(f"3. Verifying patient ID...")
            verification_success, verification_msg = self.env.verify_patient(room)
            print(f"   {verification_msg}")
            
            # 4. If staff not available, alert staff
            if not staff_available and not verified:
                print(f"4. Staff alert needed...")
                alert_msg = self.env.alert_staff(room)
                print(f"   {alert_msg}")
                
                # Try verification again after alert (simulating staff arrival)
                print(f"   Retrying verification after alert...")
                verification_success, verification_msg = self.env.verify_patient(room)
                print(f"   {verification_msg}")
            
            # 5. Deliver medicine if patient is verified
            if verification_success:
                print(f"5. Delivering medicine...")
                delivery_success, delivery_msg = self.env.deliver_medicine(room)
                print(f"   {delivery_msg}")
                
                if delivery_success:
                    print(f"   SUCCESS: Medicine delivered to room {room}")
                else:
                    print(f"   FAILED: Could not deliver medicine to room {room}")
            else:
                print(f"5. Cannot deliver - patient verification failed")
            
            print(f"\nCurrent robot inventory: {self.env.robot_inventory}")
        
        # Final System Check
        print("\n" + "=" * 70)
        print("DELIVERY COMPLETION SUMMARY")
        print("=" * 70)
        
        self.env.display_state("\nFINAL HOSPITAL STATE")
        
        # Delivery Statistics
        delivered_count = sum(1 for _, _, _, delivered, _ in self.env.rooms.values() if delivered)
        total_rooms = len(self.env.rooms)
        
        print("\nDELIVERY STATISTICS:")
        print("-" * 40)
        print(f"Total rooms to deliver: {total_rooms}")
        print(f"Successfully delivered: {delivered_count}")
        print(f"Pending deliveries: {total_rooms - delivered_count}")
        
        if self.env.delivery_log:
            print("\nDELIVERY LOG:")
            print("-" * 40)
            for log_entry in self.env.delivery_log:
                print(f"  {log_entry}")
        
        print("\nMEDICINE STORAGE STATUS (after deliveries):")
        print("-" * 40)
        for medicine, quantity in self.env.storage.items():
            print(f"  {medicine}: {quantity} units")
        
        print(f"\nRobot final inventory: {self.env.robot_inventory}")


# Main execution
hospital_env = Environment()
delivery_agent = GoalBasedAgent(hospital_env)
delivery_agent.run_agent()
```

```python
# Task 6
class Environment:
    def __init__(self):
        # 3x3 grid representation'F' = Fire, 'S' = Safe
        self.grid = { 'a': 'S', 'b': 'S', 'c': 'F', 'd': 'S', 'e': 'F', 'f': 'S', 'g': 'S', 'h': 'S', 'j': 'F'}
        
        self.room_positions = {
            'a': (0, 0),  # Top-left
            'b': (0, 1),  # Top-middle
            'c': (0, 2),  # Top-right
            'd': (1, 0),  # Middle-left
            'e': (1, 1),  # Center
            'f': (1, 2),  # Middle-right
            'g': (2, 0),  # Bottom-left
            'h': (2, 1),  # Bottom-middle
            'j': (2, 2),  # Bottom-right
        }
        
        # Predefined path for the robot
        self.path = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']
        self.fire_extinguished_count = 0
        self.total_moves = 0
    
    def get_percept(self, room):
        return self.grid.get(room, 'Unknown')
    
    def extinguish_fire(self, room):        # Update Room Status
        if self.grid[room] == 'F':
            self.grid[room] = 'S'
            self.fire_extinguished_count += 1
            return True
        return False
    
    def display_environment(self, robot_position=None):
        print("\n" + "=" * 40)
        print("CURRENT BUILDING STATUS")
        print("=" * 40)
        
        # Create a 3x3 grid display
        grid_display = [[' ' for _ in range(3)] for _ in range(3)]
        
        # Fill the grid with room statuses
        for room, (row, col) in self.room_positions.items():
            if robot_position and room == robot_position:
                grid_display[row][col] = 'ðŸ¤–'  # Robot emoji
            else:
                grid_display[row][col] = 'ðŸ”¥' if self.grid[room] == 'F' else 'ðŸŸ¢'
        
        # Print the grid
        print("\n   +---+---+---+")
        for i in range(3):
            row_display = "   |"
            for j in range(3):
                row_display += f" {grid_display[i][j]} |"
            print(row_display)
            print("   +---+---+---+")
        
        # Print legend
        print("\nLEGEND:")
        print("  ðŸ¤– = Robot position")
        print("  ðŸ”¥ = Room with fire")
        print("  ðŸŸ¢ = Safe room")
        
        # Print statistics
        print(f"\nSTATISTICS:")
        print(f"  Fires extinguished: {self.fire_extinguished_count}")
        print(f"  Total moves: {self.total_moves}")
        
        # Print room status details
        print(f"\nDETAILED ROOM STATUS:")
        for room in ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']:
            status = "FIRE ðŸ”¥" if self.grid[room] == 'F' else "SAFE ðŸŸ¢"
            robot_indicator = " (ðŸ¤– Robot Here)" if robot_position == room else ""
            print(f"  Room {room}: {status}{robot_indicator}")


class FireFightingBotAgent:
    def __init__(self, environment):
        self.env = environment
        self.current_room = None
        self.path_index = 0
        self.actions_log = []
        self.mission_complete = False
    
    def act(self, room):
        self.current_room = room
        self.env.total_moves += 1
        
        print(f"\n{'='*50}")
        print(f"ACTION: Processing Room '{room}'")
        print(f"{'='*50}")
        
        # Get current percept (room status)
        room_status = self.env.get_percept(room)
        action_result = ""
        
        # Reflexive behavior: Check for fire and act accordingly
        if room_status == 'F':
            print(f"ðŸ”¥ FIRE DETECTED in Room '{room}'!")
            if self.env.extinguish_fire(room):
                action_result = f"Extinguished fire in room '{room}'"
                self.actions_log.append(f"Room {room}: Extinguished fire ðŸ”¥â†’ðŸŸ¢")
                print(f"âœ… SUCCESS: Fire extinguished in Room '{room}'!")
            else:
                action_result = f"Failed to extinguish fire in room '{room}'"
                self.actions_log.append(f"Room {room}: Failed to extinguish")
        elif room_status == 'S':
            action_result = f"Room '{room}' is safe, moving on"
            self.actions_log.append(f"Room {room}: Already safe, moved on")
            print(f"âœ… Room '{room}' is safe. No action needed.")
        else:
            action_result = f"Unknown status for room '{room}'"
            self.actions_log.append(f"Room {room}: Unknown status")
        
        # Display environment after action
        self.env.display_environment(robot_position=room)
        
        return action_result
    
    def run_agent(self):
        print("=" * 60)
        print("FIREFIGHTING ROBOT DEPLOYMENT - MISSION START")
        print("=" * 60)
        
        # Display initial state
        print("\nINITIAL BUILDING STATUS (Before Robot Action):")
        self.env.display_environment()
        
        print("\n" + "=" * 60)
        print("ROBOT EXECUTION SEQUENCE")
        print("=" * 60)
        
        # Follow the predefined path
        for step, room in enumerate(self.env.path, 1):
            print(f"\n STEP {step}: Moving to Room '{room}'")
            print(f"   Path: {' â†’ '.join(self.env.path[:step])}")
            
            # Perform action for current room
            result = self.act(room)
            print(f"   Result: {result}")
        
        # Mission complete
        self.mission_complete = True
        print("\n" + "=" * 60)
        print("MISSION COMPLETE - ALL ROOMS PROCESSED")
        print("=" * 60)
        
        # Generate final report
        self._generate_final_report()
    
    def _generate_final_report(self):
        print("\n" + "=" * 60)
        print("FINAL MISSION REPORT")
        print("=" * 60)
        
        # Final environment status
        print("\nFINAL BUILDING STATUS:")
        self.env.display_environment()
        
        # Mission statistics
        total_fires = sum(1 for status in self.env.grid.values() if status == 'F')
        total_safe = sum(1 for status in self.env.grid.values() if status == 'S')
        
        print("\nMISSION STATISTICS:")
        print("-" * 40)
        print(f"âœ“ Total rooms processed: {len(self.env.path)}")
        print(f"âœ“ Initial fires: 3")
        print(f"âœ“ Fires extinguished: {self.env.fire_extinguished_count}")
        print(f"âœ“ Remaining fires: {total_fires}")
        print(f"âœ“ Safe rooms: {total_safe}")
        print(f"âœ“ Total robot moves: {self.env.total_moves}")
        
        # Action log
        print("\nACTION LOG:")
        print("-" * 40)
        for i, action in enumerate(self.actions_log, 1):
            print(f"{i:2}. {action}")
        
        # Mission success check
        if total_fires == 0:
            print("\n MISSION SUCCESS: All fires have been extinguished!")
            print("   The building is now completely safe.")
        else:
            print(f"\nâš ï¸  MISSION INCOMPLETE: {total_fires} fire(s) remain.")
            fire_rooms = [room for room, status in self.env.grid.items() if status == 'F']
            print(f"   Rooms with remaining fire: {', '.join(fire_rooms)}")
        
        print("\n" + "=" * 60)
        print("ROBOT SHUTTING DOWN - MISSION END")
        print("=" * 60)


# Main execution
building_env = Environment()
firefighter_bot = FireFightingBotAgent(building_env)
firefighter_bot.run_agent()
```
